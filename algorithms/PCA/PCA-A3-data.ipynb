# Principal Component Analysis (PCA)
### Dataset: [A3-data.txt](../../input/A3-data.txt)

<a id="1"></a>
## Import modules

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

<a id="2"></a>
## Loading dataset

```python
filename = '../../input/A3-data.txt'
df = pd.read_csv(filename, delimiter=',', header=0)
print(df)
```

<a id="3"></a>
## Standardize the data

```python
features = ['x', 'y', 'z', 't']
target = ['class']

x = df.loc[:, features].values

y = df.loc[:, target].values

x = StandardScaler().fit_transform(x)
```

<a id="4"></a>
## PCA Projection to 2D

```python
pca = PCA(n_components=2)
principal_components = pca.fit_transform(x)
principal_df = pd.DataFrame(data = principal_components, columns = ['principal component 1', 'principal component 2'])


print("original shape:   ", x.shape)
print("transformed shape:", principal_components.shape)

print(principal_df)

final_df = pd.concat([principal_df, df[['class']]], axis = 1)

print(final_df)

print("Synthetic A3 Dataset - Unique Classes:", df['class'].unique())
print("Synthetic A3 Dataset - Shape of PCA Data:", df.shape)
```

<a id="5"></a>
## Explained Variance

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

print(pca.explained_variance_)
print(pca.explained_variance_ratio_)

# not sure if this is needed, as of the document requirement

# the explained variance ratio for each principal component tells us the proportion
# of the dataset's variance that lies along the axis of each principal component.
# this plot is particularly useful to understand the contribution of each component
# to the overall variance.

# plot the explained variance ratio for each principal component
plt.figure(figsize=(8, 6))
# here, we create a line plot with markers at each point.
# each point represents the explained variance ratio of a principal component.
plt.plot(pca.explained_variance_ratio_, marker='o')
plt.xlabel('Principal component')  # label for the x-axis.
plt.ylabel('Explained variance ratio')  # label for the y-axis.
plt.title('Explained variance ratio by principal component')  # title of the plot.
# here, we set the x-ticks to correspond to each principal component.
plt.xticks(np.arange(len(pca.explained_variance_ratio_)), [f"pc{i+1}" for i in range(len(pca.explained_variance_ratio_))])
plt.grid(True)  # adding a grid for better readability.
plt.show()

# the cumulative explained variance gives us the total amount of variance we'd capture
# by selecting a given number of components. this plot helps in deciding how many
# components to keep while reducing dimensionality, aiming for a number where the
# cumulative explained variance reaches a point of diminishing returns (the "elbow").

# calculate and plot the cumulative explained variance
cumulative_variances = np.cumsum(pca.explained_variance_ratio_)
plt.figure(figsize=(8, 6))
# we plot a line plot again, this time showing the cumulative variance explained
# by the first n principal components.
plt.plot(cumulative_variances, marker='o')
plt.xlabel('Number of components')  # label for the x-axis.
plt.ylabel('Cumulative explained variance')  # label for the y-axis.
plt.title('Cumulative explained variance by number of components')  # title of the plot.
# we set the x-ticks to correspond to the cumulative number of components considered.
plt.xticks(np.arange(len(pca.explained_variance_ratio_)), [f"{i+1}" for i in range(len(pca.explained_variance_ratio_))])
plt.grid(True)  # adding a grid for better readability.
plt.show()
```

<a id="6"></a>
## Visualize 2D Projection

```python
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 component PCA', fontsize = 20)

targets = sorted(df['class'].unique())
print(targets)

colors = ['r', 'g', 'b', 'y', 'm', 'c']
for target, color in zip(targets,colors):
    indicesToKeep = final_df['class'] == target
    ax.scatter(final_df.loc[indicesToKeep, 'principal component 1']
               , final_df.loc[indicesToKeep, 'principal component 2']
               , c = color
               , s = 50)
ax.legend(targets)
ax.grid()
```
